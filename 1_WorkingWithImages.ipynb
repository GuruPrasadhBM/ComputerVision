```python
from skimage.io import imread, imshow
from skimage.filters import prewitt_h, prewitt_v

grayImage = imread("/content/drive/My Drive/Colab Notebooks/ComputerVision/lena.jpg", as_gray=True)

#Calculating horizontal edges
edges_prewitt_horizontal = prewitt_h(grayImage)
```

```python
#Calculating vertical edges using prewitt kernel
edges_prewitt_vertical = prewitt_v(grayImage)
```

```python
imshow(edges_prewitt_vertical, cmap='gray')
```

```python
imshow(edges_prewitt_horizontal, cmap='gray')
```
